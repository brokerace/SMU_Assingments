```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
player_count = purchase_data["SN"].nunique()
player_total = pd.DataFrame({"Total" : [player_count]})
player_total
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Finding Unique
unique_items_count = purchase_data["Item ID"].nunique()

#Calc Values
purchases = purchase_data["Price"].count()
revenue = purchase_data["Price"].sum()
average_price = round(revenue/purchases, 2)

#Set Output
purchasing_analysis = pd.DataFrame({
    "Number of Unique Items": [unique_items_count],
    "Average Price" : [average_price],
    "Number of Purchases" : [purchases],
    "Total Revenue" : [revenue]
})

#Format Output
purchasing_analysis["Average Price"] = purchasing_analysis["Average Price"].map(('${0:,.2f}').format)
purchasing_analysis["Number of Purchases"] = purchasing_analysis["Number of Purchases"].map(('{0:,.2f}').format)
purchasing_analysis["Total Revenue"] = purchasing_analysis["Total Revenue"].map(('${0:,.2f}').format)
purchasing_analysis
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
users = purchase_data.drop_duplicates(["SN"])
df_users = pd.DataFrame(users)
gender = df_users["Gender"].value_counts()
total = df_users["Gender"].count()
gender_percent = gender/total

df_gender = pd.DataFrame({
    "Gender": ["Male", "Female", "Other / Non-Disclosed"],
    "Total Count": gender,
    "Percentage of Players": gender_percent
})

df_gender["Percentage of Players"] = df_gender["Percentage of Players"].map(('{:.2%}').format)
df_gender = df_gender.set_index("Gender")
df_gender
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#By Gender
gender = purchase_data.groupby(["Gender"])

#Get Values
purch_count = gender["SN"].count()
purch_price = gender["Price"].mean()
purch_value = gender["Price"].sum()

#Drop Dupes
duplicates = purchase_data.drop_duplicates(subset = 'SN', keep = 'first')
grouped_dup = duplicates.groupby(["Gender"])

#Calc Avg
purch_avg = (gender["Price"].sum() / grouped_dup["SN"].count())

#Create Data Frame
purch_anal_gen = pd.DataFrame({"Purchase Count": purch_count,
                              "Average Purchase Price": purch_price,
                              "Total Purchase Value": purch_value,
                              "Avg Total Per Person": purch_avg})

#Format Output - Ugh, so many typing errors
purch_anal_gen["Average Purchase Price"] = purch_anal_gen["Average Purchase Price"].map(('${:.2f}').format)
purch_anal_gen["Total Purchase Value"] = purch_anal_gen["Total Purchase Value"].map(('${:.2f}').format)
purch_anal_gen["Avg Total Per Person"] = purch_anal_gen["Avg Total Per Person"].map(('${:.2f}').format)
purch_anal_genn = purch_anal_gen[["Purchase Count", "Average Purchase Price", 
                                       "Total Purchase Value", "Avg Total Per Person"]]
purch_anal_gen
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Setup Bins
persons_df = pd.DataFrame(purchase_data.groupby("SN").first())
age_bins = [0,9,14,19,24,29,34,39,999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
age_demographics = pd.DataFrame({
    "Age Range": ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"],
})

#Setup DataFrame and Format
age_demographics = age_demographics.set_index("Age Range")
age_dem = pd.cut(persons_df["Age"], age_bins, labels = group_names)
age_demographics["Total Count"] = age_dem.value_counts()
age_totals_raw = age_demographics["Total Count"]
number_ages = persons_df["Age"].count()
age_demographics["Percentage of Players"] = age_totals_raw/number_ages
age_demographics["Percentage of Players"] = age_demographics["Percentage of Players"].map(('{:.2%}').format)

age_demographics
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Setting Up bins
temp_age = purchase_data
age1_bins = [0,9,14,19,24,29,34,39,999]
age_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#Setting Up DataFrame
temp_age["Age Range"] = pd.cut(purchase_data["Age"], age1_bins, labels = age_names)


#Defining Values
age_purchase = temp_age[["Age Range", "Price"]].groupby("Age Range").count()
age_purchase["Avg Purchase Price"] = temp_age[["Age Range", "Price"]].groupby("Age Range").mean()
age_purchase["Total Purchase Value"] = temp_age[["Age Range", "Price"]].groupby("Age Range").sum()
age_df = temp_age.groupby("SN")
age_sum = age_df.sum()
age_range = age_df["Age Range"].first()
age_per = pd.concat([age_sum, age_range], 1).reset_index()

age_purchase["Average Total Per Person"] = age_per[["Age Range", "Price"]].groupby("Age Range").mean()

#Format
age_purchase = age_purchase.rename(columns = {"Price" : "Purchase Count"})
age_purchase["Avg Purchase Price"] = age_purchase["Avg Purchase Price"].map(('${:.2f}').format)
age_purchase["Total Purchase Value"] = age_purchase["Total Purchase Value"].map(('${:,.2f}').format)
age_purchase["Average Total Per Person"] = age_purchase["Average Total Per Person"].map(('${:.2f}').format)
age_purchase.head()
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
top_spenders = purchase_data[["SN", "Price"]].groupby("SN").count()

#Setting Values
top_spenders["Avg Purchase Price"] = purchase_data[["SN", "Price"]].groupby("SN").mean()
top_spenders["Total Purchase Value"] = purchase_data[["SN", "Price"]].groupby("SN").sum()

top_spenders = top_spenders.rename(columns = {"Price" : "Purchase Count"})
top_spenders = top_spenders.sort_values("Purchase Count", ascending = False)

#Format
top_spenders["Avg Purchase Price"] = top_spenders["Avg Purchase Price"].map(('${:.2f}').format)
top_spenders["Total Purchase Value"] = top_spenders["Total Purchase Value"].map(('${:,.2f}').format)
top_spenders
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
popular = purchase_data[["Item ID", "Item Name"]].groupby("Item ID").first()
popular["Purchase Count"] = purchase_data[["Price", "Item ID"]].groupby("Item ID").count()
popular["Item Price"] = purchase_data[["Price", "Item ID"]].groupby("Item ID").first()
popular["Total Revenue"] = purchase_data[["Price", "Item ID"]].groupby("Item ID").sum()

popular = popular.groupby(["Item ID", "Item Name"]).sum()
popular_df = popular.sort_values(["Purchase Count"], ascending = False)
popular_df["Item Price"] = popular_df["Item Price"].map(('${:.2f}').format)
popular_df["Total Revenue"] = popular_df["Total Revenue"].map(('${:.2f}').format)
popular_df.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
profitable = pd.DataFrame(popular.sort_values(["Total Revenue"], ascending = False))
profitable["Item Price"] = profitable["Item Price"].map(('${:.2f}').format)
profitable["Total Revenue"] = profitable["Total Revenue"].map(('${:.2f}').format)
profitable.head()
```
